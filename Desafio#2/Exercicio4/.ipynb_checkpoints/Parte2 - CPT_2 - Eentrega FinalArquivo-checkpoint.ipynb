## Imports e organização dos dados

```python
#Análise e manipulação de dados
import pandas as pd

#DataFrames
train = pd.read_csv('X_trainCharlotte.csv')
trainReview = pd.read_csv('reviewsTrainCharlotte.csv')
test = pd.read_csv('X_testCharlotte.csv')
testReview = pd.read_csv('reviewsTestCharlotte.csv')
```

## Primeira estatégia - Resultados da Exploração de Dados 

A primeira estratégia analisada será a utilização dos resultados obtidos na exploração de dados para classificar os estabelecimentos. A hipótese obtida ao fim da exploração foi a seguinte:

"Estabelecimentos abertos, pertencentes as categorias ['Shopping','Beauty & Spas','Food','Restaurants','Health & Medical'], com horarios de funcionamento na sexta-feira de [8h,9h,10h] e que não estão geograficamente muito isolados tem uma maior chance de serem considerados populares."

Para a classificação será testada a técnica de classificação Árvore de Decisão.  

Porém antes que seja possivel utilizar os algoritmos de classificação é necessário organizar os dados de forma que a aplicação destes seja possivel. Primeiramente os estabelecimentos serão agrupados de acordo com sua proximidade geográfica.

Para isso será utilizado o algoritmo de agrupamento DBSCAN. Todos os estabelecimentos serão agrupados de acordo com suas latitudes e longitudes e os rótudos dos grupos criados serão adicioanos em uma nova coluna do dataframe chamada 'clusteredCoordinates'.

```python
from sklearn.cluster import DBSCAN

dbscanConfig = DBSCAN(eps=0.02, min_samples=1)
dbscanResults = dbscanConfig.fit(train[["longitude","latitude"]])
max(dbscanResults.labels_)
```

```python
train["clusteredCoordinates"] = dbscanResults.labels_
train
```

```python
dbscanConfig = DBSCAN(eps=0.02, min_samples=1)
dbscanResults = dbscanConfig.fit(test[["longitude","latitude"]])
max(dbscanResults.labels_)
```

```python
test["clusteredCoordinates"] = dbscanResults.labels_
test
```

Em seguida os dados serão classificados de acordo com as categorias mais populares. Os estabelecimentos que apresentam a categoria 'Food' serão classificados com o rótulo 1, os que apresentam a categoria 'Shopping' serão classificados com o rótulo 2, os que apresentam a categoria 'Beauty & Spas' serão classificados com o rótulo 3, os que apresentam a categoria 'Restaurants' serão classificados com o rótulo 4, os que apresentam a categoria 'Healty & Medical' serão classificados com o rótulo 5 e por fim aqueles que não apresentam nenhuma das categorias serão classificados com o rótulo 0.

```python
train["HavePopCateg"] = 0
i = 0
while i < len(train):
    try: # Necessário pois algumas descricoes são NaN
        if((train["categories"][i].find("Food") != -1)):
            train.loc[i,"HavePopCateg"] = 1
        elif((train["categories"][i].find("Shopping") != -1)):
            train.loc[i,"HavePopCateg"] = 2
        elif((train["categories"][i].find("Beauty & Spas") != -1)):
            train.loc[i,"HavePopCateg"] = 3
        elif((train["categories"][i].find("Restaurants") != -1)):
            train.loc[i,"HavePopCateg"] = 4
        elif((train["categories"][i].find("Health & Medical") != -1)):
            train.loc[i,"HavePopCateg"] = 5
        else:
            train.loc[i,"HavePopCateg"] = 0
    except:
        train.loc[i,"HavePopCateg"] = 0
    i+=1
train
```

```python
test["HavePopCateg"] = 0
i = 0
while i < len(test):
    try: # Necessário pois algumas descricoes são NaN
        if((test["categories"][i].find("Food") != -1)):
            test.loc[i,"HavePopCateg"] = 1
        elif((test["categories"][i].find("Shopping") != -1)):
            test.loc[i,"HavePopCateg"] = 2
        elif((test["categories"][i].find("Beauty & Spas") != -1)):
            test.loc[i,"HavePopCateg"] = 3
        elif((test["categories"][i].find("Restaurants") != -1)):
            test.loc[i,"HavePopCateg"] = 4
        elif((test["categories"][i].find("Health & Medical") != -1)):
            test.loc[i,"HavePopCateg"] = 5
        else:
            test.loc[i,"HavePopCateg"] = 0
    except:
        test.loc[i,"HavePopCateg"] = 0
    i+=1
test
```

Após isso os dados serão classificados segundo seus horarios de funcionamento nas sextas feiras. Para isso o campo 'hours' será interpretado como um JSON para que apenas as sextas feiras sejam classificadas, e com isso os horarios de funcionamento serão calculados.

```python
import json

train["HoursOpenOnFriday"] = 0
i = 0
while i < len(train):
    hour = train.loc[i,"hours"]
    if(hour == hour):
        hour = hour.replace("'", "\"")
        hourJson = json.loads(hour)
        if("Friday" in hourJson):
            friday = hourJson['Friday']
            temp = []
            for num in friday.split("-"):
                temp.append(num.split(":")[0])
            time = int(temp[1])-int(temp[0])
            if(time < 0):
                time = time*(-1)
                time = 24 - time
            elif(time == 0):
                time = 24
        train.loc[i,"HoursOpenOnFriday"] = time
    i+=1
train
```

```python
test["HoursOpenOnFriday"] = 0
i = 0
while i < len(test):
    hour = test.loc[i,"hours"]
    if(hour == hour):
        hour = hour.replace("'", "\"")
        hourJson = json.loads(hour)
        if("Friday" in hourJson):
            friday = hourJson['Friday']
            temp = []
            for num in friday.split("-"):
                temp.append(num.split(":")[0])
            time = int(temp[1])-int(temp[0])
            if(time < 0):
                time = time*(-1)
                time = 24 - time
            elif(time == 0):
                time = 24
        test.loc[i,"HoursOpenOnFriday"] = time
    i+=1
test
```

Por fim os campos criados anteriormente serão utilizados para a categorização com o algoritmo de Árvore de Decisão juntamente com o campo 'is_open' que classifica os estabelecimentos que estão abertos.  

```python
#Arvore de desisao
from sklearn import tree
from sklearn.model_selection import train_test_split

pd_train, pd_test = train_test_split(train, test_size=0.7, random_state=13)

X = pd_train[["is_open","clusteredCoordinates","HavePopCateg","HoursOpenOnFriday"]]
y = pd_train['popular']

clf = tree.DecisionTreeClassifier()
clf.fit(X, y)

X_test = pd_test[["is_open","clusteredCoordinates","HavePopCateg","HoursOpenOnFriday"]]
y_test = pd_test['popular']

clf.score(X_test, y_test)
```

```python
test['popular'] = clf.predict(test[["is_open","clusteredCoordinates","HavePopCateg","HoursOpenOnFriday"]])
test[['business_id','popular']].to_csv('ResultadosPrimeiraEstrategia.csv', index=False)
```

Os resultados alcançados foram de aproximadamente 73% de acerto. Como os resultados não são muito acertivos, provalvemente esta estratégia não é muito efetiva para a classificação dos estabelecimentos.

## Segunda estatégia - Análise de Sentimento das avaliações

A segunda estratégia que será analisada é a utilização de algoritmos de Análise de Sentimentos para classificar os textos das reviews dos estabelecimentos como reviews positivas e reviews negativas.

Para a analise será utilizada o pacote nltk para Python, mais especificamente sua função 'SentimentIntesityAnalyzer'.

Será utilizado o campo 'compound' resultante de analise de sentimentos. Este campo é um valor numério no intervalo [-1:1] onde valores negativos indicam avaliações negativas e valores positivos indicam avaliações positivas. Será adicionada uma nova coluna nos dados das avaliações onde o valor 1 significa que a avaliação é positiva e o valor 0 significa que a avaliação é negativa. 

É importante ressaltar que esta analise não é um processo barato ou rápido. Para realizar a analise de todas as mais de 260.000 avaliações foram despedidas mais de 2h30, desta forma se tornou inviavel sua execução para cada teste. Para isto os resultados foram salvos em um novo arquivo csv.

```python
#Analise de sentimento
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

# CODIGO EXTREMAMENTE LENTO (APROX 2h30 DE EXECUCAO)
for review in trainReview['text']:
    if(sia.polarity_scores(review)["compound"] > 0):
        trainReview.loc[trainReview["text"] == review, "IsPositive"] = 1
    else:
        trainReview.loc[trainReview["text"] == review, "IsPositive"] = 0

trainReview.to_csv('reviewsWithSentimentAnalysis.csv')
```

```python
for review in testReview['text']:
    if(sia.polarity_scores(review)["compound"] > 0):
        testReview.loc[testReview["text"] == review, "IsPositive"] = 1
    else:
        testReview.loc[testReview["text"] == review, "IsPositive"] = 0

testReview.to_csv('TestReviewsWithSentimentAnalysis.csv')
```

Em seguida as reviews serão agrupadas através da chave 'business_id'. Os valores a respeito das avaliações serem positivas ou negativas serão somadas para cada estabelecimentos em uma nova coluna chamada 'IsPositive', visto que a soma destes valores apenas leva em consideração a quantidade de avaliações positivas recebidas por cada estabelecimento.

```python
trainReviewWithSentimetalAnalysis = pd.read_csv('reviewsWithSentimentAnalysis.csv')
trainReviewsPositives = trainReviewWithSentimetalAnalysis.groupby('business_id').sum()[['IsPositive']]
trainReviewsPositives
```

```python
testReviewWithSentimetalAnalysis = pd.read_csv('TestReviewsWithSentimentAnalysis.csv')
testReviewsPositives = testReviewWithSentimetalAnalysis.groupby('business_id').sum()[['IsPositive']]
testReviewsPositives
```

É possivel chegar nos valores para a quantidade de avaliações negativas para cada estabelecimento ao inverter os 0s e 1s do dataframe de avaliações, agrupar por 'business_id' e somar uma vez mais.

```python
trainReviewWithSentimetalAnalysis.loc[trainReviewWithSentimetalAnalysis["IsPositive"] == 0, "IsNegative"] = 1
trainReviewWithSentimetalAnalysis.loc[trainReviewWithSentimetalAnalysis["IsPositive"] == 1, "IsNegative"] = 0
trainReviewsNegatives = trainReviewWithSentimetalAnalysis.groupby('business_id').sum()[['IsNegative']]
trainReviewsNegatives
```

```python
testReviewWithSentimetalAnalysis.loc[testReviewWithSentimetalAnalysis["IsPositive"] == 0, "IsNegative"] = 1
testReviewWithSentimetalAnalysis.loc[testReviewWithSentimetalAnalysis["IsPositive"] == 1, "IsNegative"] = 0
testReviewsNegatives = testReviewWithSentimetalAnalysis.groupby('business_id').sum()[['IsNegative']]
testReviewsNegatives
```

Após isso é realizado um join dos dataframes das quantidade de reviews (positivas e negativas) com o dataframa dos estabelecimentos utilizando a chave 'business_id'.

```python
trainJoin = pd.merge(train, trainReviewsPositives, on="business_id")
trainJoin = pd.merge(trainJoin, trainReviewsNegatives, on="business_id")
trainJoin
```

```python
testJoin = pd.merge(test, testReviewsPositives, on="business_id")
testJoin = pd.merge(testJoin, testReviewsNegatives, on="business_id")
testJoin
```

Por fim três testes de classifição são feitos, um utilizando somente a quantide de reviews negativas, outro somente a quantidade de reviews positivas e por fim ambas juntas.

```python
join_train, join_test = train_test_split(trainJoin, test_size=0.7, random_state=13)

X = join_train[["IsNegative"]]
y = join_train['popular']

clf = tree.DecisionTreeClassifier()
clf.fit(X, y)

X_test = join_test[["IsNegative"]]
y_test = join_test['popular']

clf.score(X_test, y_test)
```

```python
join_train, join_test = train_test_split(trainJoin, test_size=0.7, random_state=13)

X = join_train[["IsPositive"]]
y = join_train['popular']

clf = tree.DecisionTreeClassifier()
clf.fit(X, y)

X_test = join_test[["IsPositive"]]
y_test = join_test['popular']

clf.score(X_test, y_test)
```

```python
join_train, join_test = train_test_split(trainJoin, test_size=0.7, random_state=13)

X = join_train[["IsPositive","IsNegative"]]
y = join_train['popular']

clf = tree.DecisionTreeClassifier()
clf.fit(X, y)

X_test = join_test[["IsPositive","IsNegative"]]
y_test = join_test['popular']

clf.score(X_test, y_test)
```

```python
testJoin['popular'] = clf.predict(testJoin[["IsPositive","IsNegative"]])
testJoin[['business_id','popular']].to_csv('ResultadosSegundaEstrategia.csv', index=False)
```

Os resultados obtidos utilizandos somente as avalições positivas não foram muito efetivos, porém os resultados obtidos com as avaliações negativas bem como ambas em conjunto (provavelmente por influência das negativas) foram muito melhores.

É possivel perceber que esta estratégia apresenta resultados muito melhores que a anterior (mais de 80% de acerto). Desta forma podemos dizer que esta é uma estratégia efeitiva para a classificação dos dados.
