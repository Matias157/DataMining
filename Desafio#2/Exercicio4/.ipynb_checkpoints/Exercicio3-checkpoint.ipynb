**3- Realize as seguintes atividades:**

**a) Utilize alguma implementação do Apriori, por exemplo, a da biblioteca mlxtend (python) ou do
pacote Arules (R).**

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

**b) Utilize os dados sobre compras de supermercado “supermercado.csv”, disponível no moodle.**

```python
import csv
from itertools import zip_longest

with open('supermercado.csv') as csvfile:
    rows = csv.reader(csvfile)
    res = [list(filter(None.__ne__, l)) for l in rows]
```

**c) Garanta que que um itemset não contenha dois itens iguais e que as regras mineradas contenham
no consequente somente itens do tipo "bottled beer" ou "canned beer"**

```python
def verifyWordInCommon(a,b):
    newa = a.split()
    newb = b.split()
    for itema in newa:
        for itemb in newb:
            if(itemb == itema):
                return(True)

def removeCommonWords(lista):
    for item in lista:
        newlist = lista
        for newitem in lista:
            if(item == newitem):
                pass
            else:
                if(verifyWordInCommon(item, newitem)):
                    newlist.remove(newitem)
    return(newlist)

for sublist in res:
    sublist = removeCommonWords(sublist)
```

**d) Execute o Apriori com as configurações: suporte mínimo = 0.001 e confiança = 0.25. O que
encontrou de interessante? Faça uma breve análise das regras e resultados encontrados.**

```python
te = TransactionEncoder()
encoded = te.fit(res).transform(res)
df = pd.DataFrame(encoded, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
frequent_itemsets
```

```python
dfRegras = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.25)
dfRegras
```

```python
plt.scatter(data = dfRegras, x='support', y='confidence', edgecolors='red')
plt.xlabel('supporte')
plt.ylabel('confiança')
plt.show()
```

```python
pd.set_option('display.max_rows', 10)
dfRegras.sort_values(by=['lift'],ascending=False)
```

Organizando os resultados a partir de seu valor de lift é possivel encontrar quais são as regras que são mais positivamente relacionadas. Fazendo uma breve análise dos dez primeiros resultados é possivel perceber que as principais regras implicam em produtos que normamente são utilizados juntos (em uma receita por exemplo), bem como produtos que são da mesma categoria (como no caso de {bottled beer, red/blush wine}→{liquor} onde todos são bebidas alcoólicas).
